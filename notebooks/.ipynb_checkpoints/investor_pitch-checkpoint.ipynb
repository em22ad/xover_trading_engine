# Investor Pitch

This notebook presents a decision‑ready view of the system for an investor:

- High‑level portfolio performance and risk
- Which sectors contributed most and **why** they perform well
- Which rules are most stable and profitable
- An interactive **rule detail lookup** tool

All analysis is based on the research artifacts generated by the engine:

- `portfolio_equity_curve.csv`
- `portfolio_used_trades.csv`
- `rule_stability.csv`
- `sector_investability.csv`

```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

plt.style.use("seaborn-v0_8")

RESEARCH_DIR = "../research"  # adjust if needed

def _load_csv(name: str) -> pd.DataFrame:
    path = os.path.join(RESEARCH_DIR, name)
    if not os.path.exists(path):
        print(f"[WARN] Missing research file: {name}")
        return pd.DataFrame()
    return pd.read_csv(path)

def load_research_artifacts():
    # 1) Equity curve
    equity = _load_csv("portfolio_equity_curve.csv")

    if not equity.empty:
        if "date" in equity.columns:
            equity["date"] = pd.to_datetime(equity["date"], errors="coerce")
            equity = equity.set_index("date").sort_index()
        else:
            first = equity.columns[0]
            equity[first] = pd.to_datetime(equity[first], errors="coerce")
            equity = equity.set_index(first).sort_index()

        if "equity" in equity.columns:
            equity["equity"] = pd.to_numeric(equity["equity"], errors="coerce")
        else:
            num_cols = equity.select_dtypes(include=["number"]).columns
            if len(num_cols) > 0:
                equity["equity"] = pd.to_numeric(equity[num_cols[0]], errors="coerce")

    # 2) Used trades
    used_trades = _load_csv("portfolio_used_trades.csv")
    if not used_trades.empty:
        for col in ["entry_date", "exit_date", "signal_date"]:
            if col in used_trades.columns:
                used_trades[col] = pd.to_datetime(used_trades[col], errors="coerce")
        if "ret" in used_trades.columns:
            used_trades["ret"] = pd.to_numeric(used_trades["ret"], errors="coerce")

    # 3) Rule stability
    stability = _load_csv("rule_stability.csv")

    # 4) Sector investability
    sectors = _load_csv("sector_investability.csv")

    return equity, used_trades, stability, sectors

def compute_basic_metrics(equity: pd.DataFrame) -> dict:
    if equity.empty:
        return {}

    if "equity" in equity.columns:
        eq = equity["equity"].astype(float)
    else:
        num_cols = equity.select_dtypes(include=["number"]).columns
        if len(num_cols) == 0:
            return {}
        eq = equity[num_cols[0]].astype(float)

    if not isinstance(eq.index, pd.DatetimeIndex):
        eq.index = pd.to_datetime(eq.index, errors="coerce")

    eq = eq.dropna()
    if eq.empty:
        return {}

    ret = eq.pct_change().dropna()

    total_return = eq.iloc[-1] / eq.iloc[0] - 1.0
    ann_factor = 252
    ann_return = (1 + total_return) ** (ann_factor / len(eq)) - 1.0 if len(eq) > 1 else np.nan
    ann_vol = ret.std() * np.sqrt(ann_factor)
    sharpe = ann_return / ann_vol if ann_vol > 0 else np.nan
    max_dd = ((eq / eq.cummax()) - 1.0).min()

    return {
        "start": eq.index.min(),
        "end": eq.index.max(),
        "total_return": total_return,
        "ann_return": ann_return,
        "ann_vol": ann_vol,
        "sharpe": sharpe,
        "max_drawdown": max_dd,
        "series": eq,
        "returns": ret,
    }

def plot_drawdown(eq: pd.Series):
    if eq.empty:
        return
    running_max = eq.cummax()
    dd = (eq / running_max) - 1.0

    plt.figure(figsize=(10, 4))
    dd.plot()
    plt.title("Drawdown (Underwater) Curve")
    plt.ylabel("Drawdown")
    plt.grid(True)
    plt.tight_layout()
    plt.show()

equity, used_trades, stability, sectors = load_research_artifacts()
metrics = compute_basic_metrics(equity)
metrics
```

## 5. Rule Detail Lookup Tool

Look up any rule by `rule_id` and see:

- Stability metrics
- Realized PnL and trade distribution
- Cumulative PnL over time

Use this live with an investor to answer:

> *“Show me exactly how this rule behaves, how often it trades, and what its drawdowns look like.”*

```python
from ipywidgets import interact, Dropdown, fixed

def rule_detail_view(rule_id: str, stability: pd.DataFrame, used_trades: pd.DataFrame):
    if stability.empty:
        print("No stability data available.")
        return

    rule_row = stability[stability["rule_id"] == rule_id]
    if rule_row.empty:
        print(f"No stability record found for rule_id = {rule_id}.")
        return

    print("\n=== RULE STABILITY SNAPSHOT ===")
    display(rule_row)

    if used_trades.empty:
        print("No trade data available for this rule.")
        return

    rule_trades = used_trades[used_trades["rule_id"] == rule_id].copy()
    if rule_trades.empty:
        print("This rule did not participate in the portfolio backtest.")
        return

    rule_trades = rule_trades.sort_values("entry_date")
    rule_trades["pnl_pct"] = rule_trades["ret"] * 100

    print("\n=== RULE TRADE SUMMARY ===")
    print(f"Number of trades in portfolio: {len(rule_trades)}")
    print(f"Average trade return: {rule_trades['ret'].mean()*100:.2f}%")
    print(f"Win rate: {(rule_trades['ret'] > 0).mean()*100:.2f}%")
    print(f"Worst trade: {rule_trades['ret'].min()*100:.2f}%")
    print(f"Best trade: {rule_trades['ret'].max()*100:.2f}%")

    display(rule_trades[["group", "ticker", "entry_date", "exit_date", "ret"]].head(10))

    plt.figure(figsize=(8, 4))
    rule_trades["ret"].hist(bins=30)
    plt.title(f"Distribution of Trade Returns for {rule_id}")
    plt.xlabel("Return")
    plt.ylabel("Count")
    plt.grid(True)
    plt.tight_layout()
    plt.show()

    rule_trades = rule_trades.sort_values("exit_date")
    rule_trades["cum_pnl"] = rule_trades["ret"].cumsum()

    plt.figure(figsize=(8, 4))
    plt.plot(rule_trades["exit_date"], rule_trades["cum_pnl"])
    plt.title(f"Cumulative PnL Over Time for {rule_id}")
    plt.xlabel("Exit Date")
    plt.ylabel("Cumulative Return")
    plt.grid(True)
    plt.tight_layout()
    plt.show()

if stability.empty:
    print("No stability data available for rule lookup.")
else:
    rule_ids = stability["rule_id"].unique()
    interact(
        rule_detail_view,
        rule_id=Dropdown(options=sorted(rule_ids), description="Rule ID"),
        stability=fixed(stability),
        used_trades=fixed(used_trades),
    )
```

```python
# ============================================================
# 7. Sector Daily Return Explorer (Enhanced)
# ============================================================

from ipywidgets import interact, Dropdown, fixed
import matplotlib.patches as patches

def compute_sector_daily_returns(trades: pd.DataFrame, sector: str) -> pd.Series:
    """
    Reimplementation of _compute_sector_daily_returns() for a single sector.
    Uses the same logic as the engine, but recomputed inside the notebook.
    """
    df = trades[trades["group"] == sector].copy()
    df = df[df["exit_date"].notna()].copy()

    if df.empty:
        return pd.Series(dtype=float)

    df["entry_date"] = pd.to_datetime(df["entry_date"])
    df["exit_date"] = pd.to_datetime(df["exit_date"])

    start = df["entry_date"].min()
    end = df["exit_date"].max()

    all_days = pd.date_range(start, end, freq="B")
    daily = pd.Series(0.0, index=all_days)

    for day in all_days:
        open_trades = df[(df["entry_date"] <= day) & (df["exit_date"] >= day)]
        if open_trades.empty:
            continue

        weight = 1.0 / len(open_trades)
        exits_today = open_trades[open_trades["exit_date"] == day]

        if not exits_today.empty:
            daily.loc[day] = (exits_today["ret"] * weight).sum()

    return daily


def find_loss_clusters(series: pd.Series, threshold=-0.002):
    """
    Identify clusters of losses where daily return < threshold.
    Returns a list of (start_date, end_date).
    """
    mask = series < threshold
    clusters = []
    in_cluster = False
    start = None

    for date, flag in mask.items():
        if flag and not in_cluster:
            in_cluster = True
            start = date
        elif not flag and in_cluster:
            in_cluster = False
            clusters.append((start, date))
    if in_cluster:
        clusters.append((start, series.index[-1]))

    return clusters


def sector_daily_return_view(sector1, sector2, sector3, used_trades):
    sectors = [sector1, sector2, sector3]
    daily_series = {}

    # ------------------------------------------------------------
    # 1. Overlay Plot with Loss Cluster Highlighting
    # ------------------------------------------------------------
    plt.figure(figsize=(14, 6))

    for sector in sectors:
        s = compute_sector_daily_returns(used_trades, sector)
        daily_series[sector] = s

        if not s.empty:
            plt.plot(s.index, s.values, label=f"{sector}")

            # Highlight loss clusters
            clusters = find_loss_clusters(s, threshold=-0.002)
            for start, end in clusters:
                plt.axvspan(start, end, color="red", alpha=0.15)

    plt.title("Daily Sector Returns (Overlay with Loss Clusters Highlighted)")
    plt.ylabel("Daily Return")
    plt.grid(True)
    plt.legend()
    plt.tight_layout()
    plt.show()

    # ------------------------------------------------------------
    # 2. Volatility Table
    # ------------------------------------------------------------
    vol_data = {
        sec: (ser.std() if not ser.empty else np.nan)
        for sec, ser in daily_series.items()
    }
    print("\n=== Sector Volatility (Std Dev of Daily Returns) ===")
    display(pd.DataFrame.from_dict(vol_data, orient="index", columns=["volatility"]))

    # ------------------------------------------------------------
    # 3. Loss Cluster Listing
    # ------------------------------------------------------------
    print("\n=== Loss Clusters (daily return < -0.20%) ===")
    for sec, ser in daily_series.items():
        if ser.empty:
            print(f"{sec}: No data")
            continue

        clusters = find_loss_clusters(ser, threshold=-0.002)
        print(f"\n{sec}:")
        if not clusters:
            print("  No significant loss clusters.")
        else:
            for start, end in clusters:
                print(f"  Loss cluster: {start.date()} → {end.date()}")

# Dropdown UI
if not used_trades.empty:
    sector_list = sorted(used_trades["group"].unique())

    interact(
        sector_daily_return_view,
        sector1=Dropdown(options=sector_list, description="Sector 1"),
        sector2=Dropdown(options=sector_list, description="Sector 2"),
        sector3=Dropdown(options=sector_list, description="Sector 3"),
        used_trades=fixed(used_trades),
    )
else:
    print("No trade data available for sector daily return analysis.")
```

```python
from ipywidgets import interact, SelectMultiple, FloatSlider, fixed
import numpy as np

# ============================================================
# 1. SCHEMA VALIDATION FOR price_data
# ============================================================

def validate_price_schema(df: pd.DataFrame) -> pd.DataFrame:
    print("🔍 Validating price_data schema...\n")

    required = ["date", "ticker", "open", "high", "low", "close"]
    print("Columns found:", df.columns.tolist())

    missing = [c for c in required if c not in df.columns]
    if missing:
        raise ValueError(f"❌ price_data is missing required columns: {missing}")

    if not np.issubdtype(df["date"].dtype, np.datetime64):
        print("Converting 'date' column to datetime...")
        df["date"] = pd.to_datetime(df["date"])

    numeric_cols = ["open", "high", "low", "close"]
    for col in numeric_cols:
        if not np.issubdtype(df[col].dtype, np.number):
            raise ValueError(f"❌ Column '{col}' must be numeric. Found dtype: {df[col].dtype}")

    if df["ticker"].isna().any():
        raise ValueError("❌ Ticker column contains NaN values.")

    print("✅ price_data schema validated successfully.\n")
    return df


# ============================================================
# 2. NORMALIZE price_data
# ============================================================

#price_data = validate_price_schema(price_data)
#price_data = price_data[["date", "ticker", "open", "high", "low", "close"]]

#print("Normalized price_data:")
#display(price_data.head())


# ============================================================
# 3. Dynamic Exit Logic (Option B)
# ============================================================

def apply_dynamic_exits(trades: pd.DataFrame,
                        max_profit: float,
                        max_loss: float,
                        price_data: pd.DataFrame) -> pd.DataFrame:
    trades = trades.copy()
    trades["exit_date_dynamic"] = trades["exit_date"]
    trades["ret_dynamic"] = trades["ret"]

    for idx, row in trades.iterrows():
        ticker = row["ticker"]
        entry = row["entry_date"]
        exit_scheduled = row["exit_date"]

        px = price_data[price_data["ticker"] == ticker]
        px = px[(px["date"] >= entry) & (px["date"] <= exit_scheduled)].sort_values("date")

        if px.empty:
            continue

        entry_price = px.iloc[0]["open"]
        px = px.copy()
        px["ret"] = (px["close"] / entry_price) - 1.0

        hit_profit = px[px["ret"] >= max_profit]
        hit_loss = px[px["ret"] <= max_loss]

        hit = None
        if not hit_profit.empty:
            hit = hit_profit.iloc[0]["date"]
        if not hit_loss.empty:
            hit = hit_loss.iloc[0]["date"] if hit is None else min(hit, hit_loss.iloc[0]["date"])

        if hit is not None:
            next_day = hit + pd.Timedelta(days=1)
            next_px = price_data[(price_data["ticker"] == ticker) & (price_data["date"] == next_day)]
            if not next_px.empty:
                exit_price = next_px.iloc[0]["open"]
                trades.at[idx, "exit_date_dynamic"] = next_day
                trades.at[idx, "ret_dynamic"] = (exit_price / entry_price) - 1.0

    return trades


# ============================================================
# 3b. Build TRUE DAILY EQUITY CURVE (Option B)
# ============================================================

def build_daily_equity_from_trades(trades: pd.DataFrame) -> pd.Series:
    """
    Build a daily equity curve from trades, respecting overlapping trades.
    Each trade's total return (ret_dynamic) is spread geometrically across its holding days.
    Portfolio daily return = equal-weight average of all open trades.
    """
    if trades.empty:
        return pd.Series(dtype=float)

    trades = trades.copy()
    trades["entry_date"] = pd.to_datetime(trades["entry_date"])
    trades["exit_date_dynamic"] = pd.to_datetime(trades["exit_date_dynamic"])

    daily_buckets = {}  # date -> list of per-trade daily returns

    for _, row in trades.iterrows():
        start = row["entry_date"]
        end = row["exit_date_dynamic"]
        R = float(row["ret_dynamic"])

        if pd.isna(start) or pd.isna(end) or end < start:
            continue

        days = (end - start).days
        if days <= 0:
            days = 1

        r_day = (1.0 + R)**(1.0 / days) - 1.0

        for d in pd.date_range(start, end, freq="D"):
            daily_buckets.setdefault(d, []).append(r_day)

    if not daily_buckets:
        return pd.Series(dtype=float)

    all_dates = sorted(daily_buckets.keys())
    daily_ret = [np.mean(daily_buckets[d]) for d in all_dates]

    daily_ret = pd.Series(daily_ret, index=pd.to_datetime(all_dates), name="daily_ret")
    equity = (1.0 + daily_ret).cumprod()
    equity.name = "equity"
    return equity


# ============================================================
# 4. Portfolio Recalculation Logic (with TRUE DAILY CAGR)
# ============================================================

def recompute_portfolio(allowed_sectors,
                        rule_threshold,
                        max_profit,
                        max_loss,
                        used_trades,
                        stability,
                        price_data):

    allowed_sectors = list(allowed_sectors)

    allowed_rules = stability[
        (stability["group"].isin(allowed_sectors)) &
        (stability["rule_quality_score"] >= rule_threshold)
    ]["rule_id"].unique()

    print(f"\nSelected sectors: {allowed_sectors}")
    print(f"Min rule_quality_score: {rule_threshold:.3f}")
    print(f"Rules passing filters: {len(allowed_rules)}")

    trades = used_trades[used_trades["rule_id"].isin(allowed_rules)].copy()
    print(f"Trades passing filters: {len(trades)}")

    if trades.empty:
        print("No trades match the selected criteria.")
        return

    trades = apply_dynamic_exits(trades, max_profit, max_loss, price_data)

    # === TRUE DAILY EQUITY CURVE (Option B) ===
    equity = build_daily_equity_from_trades(trades)
    if equity.empty:
        print("No equity curve could be built from trades.")
        return

    # Map daily equity back to trades: equity at each trade's dynamic exit
    trades["equity"] = trades["exit_date_dynamic"].map(
        lambda d: equity.loc[equity.index <= d].iloc[-1]
        if (equity.index <= d).any()
        else np.nan
    )

    plt.figure(figsize=(10, 5))
    plt.plot(equity.index, equity.values)
    plt.title("Reconstructed Portfolio Equity Curve (Interactive Sandbox)")
    plt.grid(True)
    plt.tight_layout()
    plt.show()

    # === TRUE CALENDAR-BASED CAGR (engine-style) ===
    total_return = equity.iloc[-1] - 1.0

    start_date = equity.index.min()
    end_date = equity.index.max()
    days = (end_date - start_date).days
    years = days / 365.0 if days > 0 else np.nan

    if years and years > 0:
        cagr = (1.0 + total_return)**(1.0 / years) - 1.0
    else:
        cagr = np.nan

    # Volatility & Sharpe (calendar-based, from daily equity)
    daily_ret = equity.pct_change().dropna()
    ann_vol = daily_ret.std() * np.sqrt(252) if not daily_ret.empty else np.nan
    sharpe = cagr / ann_vol if ann_vol and ann_vol > 0 and not np.isnan(cagr) else np.nan

    print("=== Portfolio Metrics (Interactive, True CAGR) ===")
    print(f"Total return: {total_return*100:.2f}%")
    print(f"CAGR: {cagr*100:.2f}%")
    print(f"Annualized volatility: {ann_vol*100:.2f}%")
    print(f"Sharpe: {sharpe:.2f}")
    print(f"Trades used: {len(trades)}")

    # ============================================================
    # === TRADE-BY-TRADE TABLE (Engine Format)
    # ============================================================

    print("\n=== TRADE-BY-TRADE TABLE ===")

    table = pd.DataFrame({
        "Rule": trades["rule_id"],
        "Peers": trades.get("peers", [""] * len(trades)),
        "Target": trades["ticker"],
        "Dir": trades.get("direction", [""] * len(trades)),
        "Start": trades["entry_date"],
        "End": trades["exit_date_dynamic"],
        "Return": trades["ret_dynamic"],
        "Equity": trades["equity"]
    })

    table["Start"] = table["Start"].dt.strftime("%Y-%m-%d")
    table["End"] = table["End"].dt.strftime("%Y-%m-%d")
    table["Return"] = (table["Return"] * 100).map("{:.2f}%".format)
    table["Equity"] = (table["Equity"] * 100).map("{:.2f}%".format)

    display(table)


sector_list = [str(s) for s in sorted(stability["group"].unique())]
max_q = float(stability["rule_quality_score"].max())
default_q = round(max_q * 0.5, 2)

print(f"Observed rule_quality_score range: 0.00 → {max_q:.3f}")

interact(
    recompute_portfolio,
    allowed_sectors=SelectMultiple(
        options=sector_list,
        value=tuple(sector_list),
        description="Allowed Sectors"
    ),
    rule_threshold=FloatSlider(
        min=0.0,
        max=max_q,
        step=0.01,
        value=default_q,
        description="Min Rule Quality"
    ),
    max_profit=FloatSlider(
        min=0.01,
        max=0.50,
        step=0.01,
        value=0.20,
        description="Max Profit Exit"
    ),
    max_loss=FloatSlider(
        min=-0.50,
        max=-0.01,
        step=0.01,
        value=-0.10,
        description="Max Loss Exit"
    ),
    used_trades=fixed(used_trades),
    stability=fixed(stability),
    price_data=fixed(price_data),
)
```
